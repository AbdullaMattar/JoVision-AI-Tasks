Train on YoloV8


```python
from ultralytics import YOLO

# Load a pre-trained fast YOLO model
model = YOLO('yolov8m.pt')

# Start training
results = model.train(data='/content/data.yaml', epochs=50, imgsz=640)
```

**Test** the model



```python
from ultralytics import YOLO
from IPython.display import display, Image

# 1. Load your newly trained model (Ultralytics saves it here by default)
model = YOLO('runs/detect/train3/weights/best.pt')

# 2. Run prediction on a test image (make sure to upload a test image to Colab)
# save=True tells it to draw the boxes and save the new image
results = model.predict(source='/content/test.jpg', save=True)
```
