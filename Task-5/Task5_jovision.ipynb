Train on YoloV8


```python
from ultralytics import YOLO

# Load the base model again
model = YOLO('yolov8m.pt')

# Re-train with heavy rotation and vertical flipping
results = model.train(
    data='data.yaml',
    epochs=75,
    imgsz=640,
    degrees=180,     # Rotate pieces
    flipud=0.5,      # 50% chance to train on upside-down pieces
    name='train_augmented_V2'
)
```

**Test** the model



```python
from ultralytics import YOLO
from IPython.display import display, Image

# 1. Load your newly trained model (Ultralytics saves it here by default)
model = YOLO('runs/detect/train_augmented_V2/weights/best.pt')

# 2. Run prediction on a test image (make sure to upload a test image to Colab)
# save=True tells it to draw the boxes and save the new image
results = model.predict(source='/content/test2_editied.png', save=True)
```
